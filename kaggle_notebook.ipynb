```python
# ---- Set your inputs here ----
INPUT_IMG  = f'{DATA_DIR}/images/0.jpg'   # path to your face/subject image
INPUT_MASK = f'{DATA_DIR}/masks/0.jpg'    # path to corresponding mask
PROMPT     = 'a photo of a person attending a formal event.'
MODEL_PATH = 'stabilityai/stable-diffusion-2-1-base'  # downloaded from HF automatically
SEED       = 42
# ------------------------------
```

```python
# Copy sample data if no custom images provided
import shutil
if not os.path.exists(INPUT_IMG):
    shutil.copy(f'{WORKING_DIR}/COW/data/images/0.jpg', INPUT_IMG)
if not os.path.exists(INPUT_MASK):
    shutil.copy(f'{WORKING_DIR}/COW/data/masks/0.jpg', INPUT_MASK)
```

```python
!python run_COW.py \
    --input_img  {INPUT_IMG} \
    --input_mask {INPUT_MASK} \
    --prompt     "{PROMPT}" \
    --model_path {MODEL_PATH} \
    --output_dir {OUTPUT_DIR} \
    --seed       {SEED}
```
